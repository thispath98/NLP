실제로 `마블의운은 여기까지이군 ㅋㅋㅋㅋ 잘가라 마블 즐거웠다`와 같은 부정적인 반응을 긍정적으로 판단하거나, 긍정적인 평을 부정적으로 판단한 경우도 있었다.

하지만 놀라운 것은 `인생 0순위 영화입니다`와 같은 애매모호한 평가도 긍정적으로 판단할 수 있었다는 점이다.

```python
movie_rank = open(movie_rank_file, 'r')
prediction = open('/content/KoBERT-nsmc/sample_pred_out.txt', 'r')

rank = movie_rank.readline().rstrip()
pred = prediction.readline().rstrip()
while rank and pred:
    print(f"평점 = {rank} / 결과 = {pred}")
    rank = movie_rank.readline().rstrip()
    pred = prediction.readline().rstrip()

movie_rank.close()
prediction.close()
```
